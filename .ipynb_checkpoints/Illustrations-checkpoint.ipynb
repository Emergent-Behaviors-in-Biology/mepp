This notebook reproduces all the illustrative examples for arXiv:1901.09673, "The Minimum Environmental Perturbation Principle."

In addition to the standard Anaconda packages, it requires the Community Simulator package from https://github.com/Emergent-Behaviors-in-Biology/community-simulator to run the simulations.

# Imports and definitions

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends import backend_pdf as bpdf
import matplotlib
import pickle
from community_simulator.usertools import MakeConsumerDynamics,MakeResourceDynamics
from community_simulator import Community
from scipy.optimize import brentq
%matplotlib inline

colors = sns.color_palette()
Rmax = 5
nR = 100

R0 = [4.8,2.85]

#Construct perturbation measures
def biotic(R1,R2,K1=R0[0],K2=R0[1]):
    return (R1-K1)**2 + (R2-K2)**2
def abiotic(R1,R2,K1=R0[0],K2=R0[1]):
    return K1*np.log(K1/R1) - (K1 - R1) + K2*np.log(K2/R2) - (K2 - R2)

#Construct dynamics
assumptions_abio = {'regulation':'independent','supply':'external','response':'type I'}
def dNdt1(N,R,params):
    return MakeConsumerDynamics(assumptions_abio)(N,R,params)
def dRdt1(N,R,params):
    return MakeResourceDynamics(assumptions_abio)(N,R,params)
dynamics_abio = [dNdt1,dRdt1]

assumptions_bio = {'regulation':'independent','supply':'self-renewing','response':'type I'}
def dNdt2(N,R,params):
    return MakeConsumerDynamics(assumptions_bio)(N,R,params)
def dRdt2(N,R,params):
    return MakeResourceDynamics(assumptions_bio)(N,R,params)
dynamics_bio = [dNdt2,dRdt2]

def dNdt3(N,R,params):
    return N*((params['gamma']/(params['c'].dot(1/R))) - params['m'])
def dRdt3(N,R,params):
    return (params['R0']-R)/params['tau'] - params['gamma']*((params['c'].T)*(((params['c'].dot(1/R))**(-2)).T)).dot(N)/R
dynamics_essential = [dNdt3,dRdt3]

#Define grid for contours
delta = 0.025
R1 = np.arange(0, Rmax, delta)
R2 = np.arange(0, Rmax, delta)
R1 = R1[1:]
R2 = R2[1:]
R1_mat, R2_mat = np.meshgrid(R1,R2)
```

# Figure 2: Self-Renewing Resources

```python
#Create parameter set
m = 2
c = np.asarray([[0.5,0.3],[0.4,0.6]])
D = np.asarray([[0,1],[1,0]])
R0 = [4.8,2.85]
params={'c':c,
        'm':m,
        'w':1,
        'g':1,
        'R0':np.asarray(R0),
        'r':1.
       }

#Initialize list to hold trajectory outputs
R_list = []
#Set initial population sizes
R0_traj = [[1,4],[4.5,4.5]]
for k in range(len(R0_traj)):
    #Simulate for each initial state
    init_state=[np.asarray([1,1]),np.asarray(R0_traj[k])]
    MyPlate = Community(init_state,dynamics_bio,params)
    t, N_traj, R_traj = MyPlate.TestWell(10,ns=100)
    R_list.append(R_traj)

#Initialize figure
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)

#Shade uninvadable region
Rcross = np.linalg.inv((c)).dot(np.asarray([m,m]))
R11 = np.linspace(0,Rcross[0],nR)
R10 = np.linspace(Rcross[0],Rmax,nR)
R21 = -c[1,0]*R11/(c[1,1]) + m/(c[1,1])
R20 = -c[0,0]*R10/(c[0,1]) + m/(c[0,1])
R10 = R10[R20>=0]
R20 = R20[R20>=0]
ax.fill_between(R10,R20,color='gray')
ax.fill_between(R11,R21,color='gray')

#Plot ZNGI's
R = np.linspace(0,Rmax,100)
ax.plot(R,m/(c[0][1])-c[0][0]*R/(c[0][1]),color=colors[0],lw=3)
ax.plot(R,m/(c[1][1])-c[1][0]*R/(c[1][1]),color=colors[1],lw=3)

#Plot trajectories
for k in range(len(R0_traj)):
    ax.quiver(R_list[k][:-1, 0], R_list[k][:-1, 1], R_list[k][1:, 0]-R_list[k][:-1, 0], 
              R_list[k][1:, 1]-R_list[k][:-1, 1], scale_units='xy', angles='xy', scale=1, color='k') 
    #ax.plot(R_list[k][:,0],R_list[k][:,1],'--',color='k',lw=1)

#Plot equilibrium point
ax.plot([R_list[0][-1,0]],[R_list[0][-1,1]],'o',color='k')

#Make contour plots
f0 = biotic(Rcross[0],Rcross[1],K1 = R0[0],K2 = R0[1])
CS = ax.contour(R1_mat, R2_mat, biotic(R1_mat, R2_mat,K1 = R0[0],K2 = R0[1]),np.arange(1,25)*f0/3,colors='k',linestyles='dotted')

ax.scatter([R0[0]],[R0[1]],s=30,color='k',marker='x')
ax.axis('square')
ax.set_xlim((0,Rmax))
ax.set_ylim((0,Rmax))
ax.set_xlabel('Resource 1', fontsize=12)
ax.set_ylabel('Resource 2', fontsize=12)
pdf = bpdf.PdfPages('MacArthur2.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()

#################

#Set up landscape plot       
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)
R = np.vstack((np.hstack((R11,R10)),np.hstack((R21,R20)))).T

#Plot boundary of uninvadable region
ax.set_xlim((0,Rmax))
ax.plot([R[nR,0],R[-1,0]],[0,0],color=colors[0],lw=3)
ax.plot([0,R[nR-1,0]],[0,0],color=colors[1],lw=3)
ax.plot([R[nR-1,0],R[nR-1,0]],[-0.1,10],'k',lw=1)

#Make Euclidean distance
Delta = R0-R
d = 0.5*(Delta[:,0]**2+Delta[:,1]**2)
ax.plot(R[:,0],d,color = 'k')

#Plot final state from trajectories
ax.plot([R_list[0][-1][0]],[0],'o',color='k')

#Format
ax.spines['bottom'].set_visible(False)
ax.set_ylim((-0.2,10))
ax.set_xlabel('Resource 1',fontsize=12)
ax.set_ylabel('Environmental Perturbation',fontsize=12)

pdf = bpdf.PdfPages('Distance_MacArthur.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

# Figure 3: Non-substitutable Resources

```python
#Create parameter set
gamma = 1
m = 1
c = np.asarray([[1,0.3],[0.3,1]])
R0 = [4.8,2.85]

params={'c':c,
        'm':m,
        'R0':np.asarray(R0),
        'tau':1.,
        'gamma':gamma
       }

#Initialize list to hold trajectory outputs
R_list = []
#Set initial population sizes
R0_traj = [[2,4],[4.5,4.5]]
N0 = [[1,1],[1,1e-2]]
T = [10,150]
ns = [60,1000]
nplot = [59,80]
for k in range(len(R0_traj)):
    #Simulate for each initial state
    init_state=[np.asarray(N0[k]),np.asarray(R0_traj[k])]
    MyPlate = Community(init_state,dynamics_essential,params)
    t, N_traj, R_traj = MyPlate.TestWell(T[k],ns=ns[k])
    R_list.append(R_traj)

#Initialize figure
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)

#Shade uninvadable region
def ZNGI(R1,cvec):
    return R1*cvec[1]*m/(gamma*R1 - m*cvec[0])
def delta(R1):
    return ZNGI(R1,c[0]) - ZNGI(R1,c[1])
R1cross = brentq(delta,1,2)
Rcross = [R1cross,ZNGI(R1cross,c[0])]
R11 = np.linspace(0.3,Rcross[0],nR)
R10 = np.linspace(Rcross[0],Rmax,nR)
R21 = ZNGI(R11,c[1])
R20 = ZNGI(R10,c[0])
R10 = R10[R20>=0]
R20 = R20[R20>=0]
ax.fill_between(R10,R20,color='gray')
ax.fill_between(R11,R21,color='gray')
ax.fill_between([0,0.3],[5,5],color='gray')

#Plot ZNGI's
R = np.linspace(0.5,Rmax,100)
ax.plot(R,ZNGI(R,c[0]),color=colors[0],lw=3)
R = np.linspace(0.3,Rmax,100)
ax.plot(R,ZNGI(R,c[1]),color=colors[1],lw=3)


#Plot trajectories
for k in range(len(R0_traj)):
    ax.quiver(R_list[k][:nplot[k], 0], R_list[k][:nplot[k], 1], R_list[k][1:nplot[k]+1, 0]-R_list[k][:nplot[k], 0], 
                  R_list[k][1:nplot[k]+1, 1]-R_list[k][:nplot[k], 1], scale_units='xy', angles='xy', scale=1, color='k', linewidths=1) 
    ax.plot([R_list[k][-1,0]],[R_list[k][-1,1]],'ok')
    #ax.plot(R_list[k][:,0],R_list[k][:,1],'--',color='k',lw=1)

#Make contour plots
f0 = abiotic(R_list[0][-1,0],R_list[0][-1,1],K1 = R0[0],K2 = R0[1])
CS = ax.contour(R1_mat, R2_mat, abiotic(R1_mat, R2_mat,K1 = R0[0],K2 = R0[1]),np.arange(1,25)*f0/3,colors='k',linestyles='dotted')

ax.scatter([R0[0]],[R0[1]],s=30,color='k',marker='x')
ax.axis('square')
ax.set_xlim((0,Rmax))
ax.set_ylim((0,Rmax))
ax.set_xlabel('Resource 1', fontsize=12)
ax.set_ylabel('Resource 2', fontsize=12)
pdf = bpdf.PdfPages('Essential.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()

#################

#Set up landscape plot       
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)
R = np.vstack((np.hstack((R11,R10)),np.hstack((R21,R20)))).T

#Plot boundary of uninvadable region
nR_new = len(R10)
ax.set_xlim((0,Rmax))
ax.plot([R[nR_new,0],R[-1,0]],[3.01,3.01],color=colors[0],lw=3)
ax.plot([0,R[nR_new-1,0]],[3.01,3.01],color=colors[1],lw=3)

#Make Euclidean distance
Delta = R0-R
d=(np.log(R0/R)*R0 - (R0-R)).sum(axis=1)
ax.plot(R[:,0],d,color = 'k')

#Plot final state from trajectories
for k in range(2):
    ax.plot([R_list[k][-1,0]],[3.01],'o',color='k')
    ax.plot([R_list[k][-1][0],R_list[k][-1][0]],[-0.1,10],color='k',lw=1)

#Format
ax.spines['bottom'].set_visible(False)
ax.set_ylim((3,4))
ax.set_xlabel('Resource 1',fontsize=12)
ax.set_ylabel('Environmental Perturbation',fontsize=12)

pdf = bpdf.PdfPages('Distance_Essential.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

# Figure 3: Environmental Feedbacks

```python
#Create parameter set
c = np.asarray([[0.5,0.3],[0.4,0.6]])
D = np.asarray([[0,1],[1,0]])
R0 = [4.8,0.9]
l=0.5
m=2
params={'c':c,
        'D':D,
        'm':m*(1-l),
        'l':l,
        'w':1,
        'g':1,
        'R0':np.asarray(R0),
        'tau':1.
       }

#Initialize list to hold trajectory outputs
R_list = []
#Set initial population sizes
R0_traj = [[1,4],[4.5,4.5]]
for k in range(len(R0_traj)):
    #Simulate for each initial state
    init_state=[np.asarray([1,1]),np.asarray(R0_traj[k])]
    MyPlate = Community(init_state,dynamics_abio,params)
    t, N_traj, R_traj = MyPlate.TestWell(20,ns=100)
    R_list.append(R_traj)

#Initialize figure
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)

#Shade uninvadable region
Rcross = np.linalg.inv((c)).dot(np.asarray([m,m]))
R11 = np.linspace(0,Rcross[0],nR)
R10 = np.linspace(Rcross[0],Rmax,nR)
R21 = -c[1,0]*R11/(c[1,1]) + m/(c[1,1])
R20 = -c[0,0]*R10/(c[0,1]) + m/(c[0,1])
R10 = R10[R20>=0]
R20 = R20[R20>=0]
ax.fill_between(R10,R20,color='gray')
ax.fill_between(R11,R21,color='gray')

#Plot ZNGI's
R = np.linspace(0,Rmax,100)
ax.plot(R,m/(c[0][1])-c[0][0]*R/(c[0][1]),color=colors[0],lw=3)
ax.plot(R,m/(c[1][1])-c[1][0]*R/(c[1][1]),color=colors[1],lw=3)

#Plot trajectories
for k in range(len(R0_traj)):
    ax.quiver(R_list[k][:-1, 0], R_list[k][:-1, 1], R_list[k][1:, 0]-R_list[k][:-1, 0], 
              R_list[k][1:, 1]-R_list[k][:-1, 1], scale_units='xy', angles='xy', scale=1, color='k') 
    #ax.plot(R_list[k][:,0],R_list[k][:,1],'--',color='k',lw=1)

#Plot equilibrium point
ax.plot([R_list[0][-1,0]],[R_list[0][-1,1]],'o',color='k')

#Make contour plots
Q = np.eye(2) - l*D
Qinv = np.linalg.inv(Q)
w_eff = np.diag(Qinv)
Qinv = Qinv - np.diag(np.diag(Qinv))
R0t = R0 + (1/w_eff)*Qinv.dot(R0-R_list[1][-1,:])
f0 = abiotic(R_list[0][-1,0],R_list[0][-1,1],K1 = R0t[0],K2 = R0t[1])
CS = ax.contour(R1_mat, R2_mat, abiotic(R1_mat, R2_mat,K1 = R0[0],K2 = R0t[1]),np.arange(1,25)*f0,colors='k',linestyles='dotted')

#Plot unperturbed point
ax.scatter([R0[0]],[R0[1]],s=30,color='k',marker='x',alpha=0.5)
ax.scatter([R0t[0]],[R0t[1]],s=30,color='k',marker='x')

ax.axis('square')
ax.set_xlim((0,Rmax))
ax.set_ylim((0,Rmax))
ax.set_xlabel('Resource 1', fontsize=12)
ax.set_ylabel('Resource 2', fontsize=12)
pdf = bpdf.PdfPages('Crossfeeding.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()

#################

#Set up landscape plot       
fig,ax=plt.subplots(figsize=(3,3))
fig.subplots_adjust(left=0.2,bottom=0.19)
R = np.vstack((np.hstack((R11,R10)),np.hstack((R21,R20)))).T

#Plot boundary of uninvadable region
ax.set_xlim((0,Rmax))
ax.plot([R[nR,0],R[-1,0]],[0,0],color=colors[0],lw=3)
ax.plot([0,R[nR-1,0]],[0,0],color=colors[1],lw=3)
ax.plot([R[nR-1,0],R[nR-1,0]],[-0.1,10],'k',lw=1)

#Make Euclidean distance
Delta = R0-R
d=(w_eff*(np.log(R0t/R)*R0t - (R0t-R))).sum(axis=1)
ax.plot(R[:,0],d,color = 'k')

#Plot final state from trajectories
ax.plot([R_list[0][-1][0]],[0],'o',color='k')

#Format
ax.spines['bottom'].set_visible(False)
ax.set_ylim((-0.2,4))
ax.set_xlabel('Resource 1',fontsize=12)
ax.set_ylabel('Environmental Perturbation',fontsize=12)

pdf = bpdf.PdfPages('Distance_crossfeeding.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```
